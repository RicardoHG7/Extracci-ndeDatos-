#Filtro 6

```python
#Este filtro nos permite ver datos en relación a una variable cuantitativa , en este caso se eligieron aquellos precios unitarios que fuesen menores a 134

#Filtro por comparacion
filtro6=df[df["PRECIO_UNITARIO"] < 134]
filtro6
```

#Filtro 7 

```python
#Este filtro nos ayuda a buscar datos en espesifico en relación a las columnas y criterios cuantitativos.

#Filtro Y
filtro7=df[(df["PRECIO_UNITARIO"] > 2000) & (df["MARGEN_UNITARIO_CALCULADO"] == "2022-11-30")]
filtro7
```

```python
#Convertir archivo filtrado a CSV
filtro7.to_csv("filtro7.csv")
#descardar archivo filtrado en csv
from google.colab import files   
files.download("filtro7.csv")
```

#Filtro 8 


```python
#Este filtro nos ayuda a buscar datos en espesifico en relación a las columnas y criterios cuantitativos.
#Filtro o
filtro8=df[(df["PRECIO_UNITARIO"] < 2000) | (df["MARGEN_UNITARIO_CALCULADO"] == "2022-11-30")]
filtro8
```

```python
#Convertir archivo filtrado a CSV
filtro8.to_csv("filtro8.csv")
#descardar archivo filtrado en csv
from google.colab import files   
files.download("filtro8.csv")
```

#Filtro 9

```python
#Este filtro es para que no se cumpla lo indicado en el codigo 
#Filtro not
filtro9=df[~(df["PRECIO_UNITARIO"] >= 2000) & ~(df["MARGEN_UNITARIO_CALCULADO"] == "2022-11-30")]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("filtro9.csv")
#descardar archivo filtrado en csv
from google.colab import files   
files.download("filtro9.csv")
```
