```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx sin indice
df = pd.read_excel('Datos.xlsx')
df
```

#CVE_CLPV      (valores: 1000 a 2000)

```python
df['CVE_CLPV'] = pd.to_numeric(df['CVE_CLPV'], errors='coerce')
df1 = df[(df['CVE_CLPV'] >= 1000) & (df['CVE_CLPV'] <= 2000)]
df1
```

```python
#Convertir archivo filtrado a CSV
df1.to_csv("df1.csv")
#descardar archivo filtrado en csv
from google.colab import files   
files.download("df1.csv")
```

#CVE_VEND     (Todas las claves excepto “5” y “4”)

```python
df2 = df[~df['CVE_VEND'].isin(['4', '5'])]
df2
```

```python
#Convertir archivo filtrado a CSV
df2.to_csv("df2.csv")
#descardar archivo filtrado en csv
from google.colab import files   
files.download("df2.csv")
```

#FECHA_ENT   (Fechas de “28/02/2022”)

```python
df3 = df[df['FECHA_ENT'] == '2022-02-28']
df3
```

```python
#Convertir archivo filtrado a CSV
df3.to_csv("df3.csv")
#descardar archivo filtrado en csv
from google.colab import files   
files.download("df3.csv")
```

#CAN_TOT       (Cantidades menores a 5951.7)   o  STATUS (“E”)
